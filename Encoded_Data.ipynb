```
import pandas as pd
import matplotlib.pyplot as plt
import pylab
import numpy as np
import matplotlib.pyplot as plt


from sklearn.neural_network import MLPRegressor
from sklearn.metrics import accuracy_score
data = pd.read_csv('data_banknote_authentication.txt', sep=",", header=None)
data.columns = ['variance','skew','curtosis','entropy','class']
labels=['variance','skew','curtosis','entropy']
X_train=data[data['class']==0]

X_test=data[data['class']==1]
X_train=X_train[labels]
X_test=X_test[labels]
print(X_train)
import keras
from sklearn import metrics
import numpy as np
import pandas as pd
from IPython.display import display, HTML 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
Xtrain=X_train.values
Xtest=X_test.values
print(Xtrain.shape[1])
model = Sequential()
model.add(Dense(50, input_dim=Xtrain.shape[1], activation='relu',use_bias=True))##input Layer
model.add(Dense(50, activation='relu'))#Hideen Layer 1 with 50 neurons and 50 output neurons(encoded)
model.add(Dense(2, activation='relu')) #Hideen Layer 2 with 50 neurons and 2 output neurons(compressed to two neurons)(encoded)
model.add(Dense(50, activation='relu'))#Bottleneck with Three neurons with outputlayer considt of 50 neurons in decoding layer
model.add(Dense(50, activation='relu'))#Bottleneck with Three neurons with outputlayer considt of 50 neurons in decoding layer
model.add(Dense(Xtrain.shape[1])) # OUt put neurons with same number of input dimension
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(Xtrain,Xtrain,verbose=1,epochs=300)
pred = model.predict(Xtest)#decoded data
pred=pred.tolist()
errors=[]
test=Xtest.tolist()
for i in range(len(pred)):
  s=0
  for j in range(4):
    s=s+(pred[i][j]-test[i][j])**2
  errors.append(s**0.5)
import statistics
mean=statistics.mean(errors)
print("Mean",mean)
std=statistics.stdev(errors)
print("standdard deviation",std)
print("max",max(errors))
print("min",min(errors))
count=0
print(Xtest)
print(pred)
for i in range(len(errors)):
  if(errors[i]>=mean+std):
    count=count+1
print("Number of test samples")
print(len(errors))
print("No of Anamolies")
print(count)



```
